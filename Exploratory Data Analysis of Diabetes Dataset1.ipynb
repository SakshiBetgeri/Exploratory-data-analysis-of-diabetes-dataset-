# Exploratory Data Analysis of Diabetes Dataset Using Python

## Objective
- Analyze healthcare dataset
- Identify factors affecting diabetes
- Perform EDA and visualization
- Build an interactive dashboard

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from ipywidgets import interact
import ipywidgets as widgets

sns.set(style="whitegrid")
```

```python
import pandas as pd

df = pd.read_csv("./diabetes.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
cols = ["Glucose", "BloodPressure", "BMI", "Insulin"]

df[cols] = df[cols].replace(0, np.nan)

df.fillna(df.median(), inplace=True)
```

```python
df.hist(figsize=(10,8))
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.countplot(x="Outcome", data=df)
plt.title("Diabetes Outcome Distribution")
plt.show()
```

```python
sns.boxplot(x="Outcome", y="Glucose", data=df)
plt.title("Glucose vs Diabetes")
plt.show()
```

```python
sns.scatterplot(x="Age", y="BMI", hue="Outcome", data=df)
plt.title("Age vs BMI")
plt.show()
```

```python
sns.kdeplot(data=df, x="Glucose", hue="Outcome", fill=True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df, hue="Outcome")
plt.show()
```

```python
import plotly.express as px
```

```python
fig = px.scatter(df, x="Glucose", y="BMI", color="Outcome",
                 title="Glucose vs BMI (Interactive)")
fig.show()
```

```python
fig = px.histogram(df, x="Age", color="Outcome",
                   title="Age Distribution")
fig.show()
```

```python
fig = px.imshow(df.corr(), text_auto=True,
                title="Correlation Heatmap")
fig.show()
```
